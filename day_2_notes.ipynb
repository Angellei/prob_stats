```python
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.stats as scs
import seaborn as sns
sns.set()
```

```python
def plot_distributions(distr, continuous, X, title, x_label, x_lim):
    title_save = '_'.join(title.lower().split())
    
    if continuous: 
        y = [distr.pdf(k) for k in X]
        plt.plot(X, y, color='r')

    else: 
        y = [distr.pmf(k) for k in X]
        plt.bar(X, y, color='r')

    plt.xlabel(x_label)    
    plt.xlim(x_lim)
    plt.title('Example: {}'.format(title))
    plt.tight_layout()
    plt.savefig('question_{}_pdf.png'.format(title_save))
    plt.show()

    y_cdf = [distr.cdf(k) for k in X]
    if continuous: 
        y = [distr.pdf(k) for k in X]
        plt.plot(X, y_cdf, color='r')

    else: 
        y = [distr.pmf(k) for k in X]
        plt.bar(X, y_cdf, color='r')

    plt.xlabel(x_label)
    plt.xlim(x_lim)
    plt.title('Example: {}'.format(title))
    plt.tight_layout()
    plt.savefig('question_{}_pdf.png'.format(title_save))
    plt.show()
```

# BINOMIAL

```python
n = 100
p = .4

continuous = False
distr = scs.binom(n, p)
X = np.arange(0, 60, 1)
title = 'Something'
x_label = 'random variable'
x_lim = (0, 60)

plot_distributions(distr, continuous, X, title, x_label, x_lim)
```

# POISSON

```python
mu = 10

continuous = False
distr = scs.poisson(mu)
X = np.arange(0, 20, 1)
title = 'Something'
x_label = 'random variable'
x_lim = (0, 20)

plot_distributions(distr, continuous, X, title, x_label, x_lim)
```

# NORMAL DISTRIBUTION

```python
mu = 10
sigma = 2

continuous = True
distr = scs.norm(mu, sigma)
X = np.arange(0, 20, .1)
title = 'Something'
x_label = 'random variable'
x_lim = (0, 20)

plot_distributions(distr, continuous, X, title, x_label, x_lim)
```

# EXPONENTIAL

```python
mu = 5

continuous = True
distr = scs.expon(0, mu)
X = np.arange(0, 20, .1)
title = 'Something'
x_label = 'random variable'
x_lim = (0, 20)

plot_distributions(distr, continuous, X, title, x_label, x_lim)
```

# Hypothesis Testing

### Example:
I toss a coin 500 times and it lands heads 263 times.   Is the coin fair?

```python
# Wrong solution num 1:

n=500
p=.5
k=263

distr = scs.binom(n, p)
prob = distr.pmf(k)
print("Prob of flipping {} heads out of {}: {:.3f}".format(k, n, prob))
```

```python
# Wrong solution num 2:

n = 500
p_fair = .5
p_bias = 263/500

fair_distr = scs.binom(n, p_fair)
bias_distr = scs.binom(n, p_bias)     

prob_fair = fair_distr.pmf(263)
prob_bias = bias_distr.pmf(263)
                       
'''P(263|fair coin)'''
print("Probability of 263 heads given a fair coin: {:.3f}".format(prob_fair))

'''P(263|biased coin)'''
print("Probability of 263 heads given a biased coin: {:.3f}".format(prob_bias))
```
